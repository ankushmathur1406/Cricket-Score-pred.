```python
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')  # Convert non-numeric values to NaN
df.dropna(subset=['runs'], inplace=True)  # Remove rows with NaN values
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
#df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
#df['wickets_left'] = 10 - df['player_dismissed']

df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = (10 - df['player_dismissed']).clip(lower=0)  # Ensures values stay within [0, 10]


#df['player_dismissed'] = pd.to_numeric(df['player_dismissed'], errors='coerce')  # Convert non-numeric values to NaN
#df.dropna(subset=['player_dismissed'], inplace=True)  # Remove rows with NaN values
#df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
#df['wickets_left'] = 10 - df['player_dismissed']
print(df.dtypes)





```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
#groups = df.groupby('match_id')

#match_ids = df['match_id'].unique()
#last_five = []
#for id in match_ids:
  #  last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
#df['last_five']=last_five
print(df['runs'].unique())  # Identify unexpected values
df = df[pd.to_numeric(df['runs'], errors='coerce').notna()]
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')  # Converts invalid values to NaN
df.dropna(subset=['runs'], inplace=True)  # Remove rows with NaN value
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []

for id in match_ids:
    last_five.extend(groups.get_group(id)['runs'].rolling(window=30).sum().values.tolist())

df['last_five'] = last_five
print(df.dtypes)
```

```python
X = final_df.drop(columns=['total_runs'])
y = final_df['total_runs']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city', 'gender'])
], remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
